# 4.2 两种画图接口

## 4.2.1 MATLAB风格

```python
plt.figure
# 创建两个子图中的第一个，设置坐标轴
# （行、列、子图编号）
plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x))

# 创建两个子图中的第二个，设置坐标轴
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
```

## 4.2.2 面向对象接口

```python
# 先创建图形网络
# ax是一个包含两个Axes对象的数组
fig, ax = plt.subplots(2)

# 在每个对象上调用plot()方法
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

## 4.3.1 调整图形：线条的颜色与风格

颜色的不同表示方法：

```python
# 标准颜色名称
plt.plot(x, np.sin(x - 0), color='blue')
# 缩写颜色代码
plt.plot(x, np.sin(x - 1), color='g')
# 范围在0~1的灰度值
plt.plot(x, np.sin(x - 2), color='0.75')
# 十六进制
plt.plot(x, np.sin(x - 3), color='#FFDD44')
# RGB元组，范围在0~1
plt.plot(x, np.sin(x - 4), color=(1.0, 0.2, 0.3))
# HTML颜色名称
plt.plot(x, np.sin(x - 5), color='chartreuse')
```

用linestyle调整线条的风格：

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted')

# 也可以用下面的简写形式
plt.plot(x, x + 4, linestyle='-') # 实线
plt.plot(x, x + 5, linestyle='--') # 虚线
plt.plot(x, x + 6, linestyle='-.') # 点划线
plt.plot(x, x + 7, linestyle=':') # 实点线
```

```python
组合使用linestyle和color编码
```

```python
plt.plot(x, x + 0, '-g') # 绿色实线
plt.plot(x, x + 1, '--c') # 青色虚线
plt.plot(x, x + 2, '-.k') # 黑色点划线
plt.plot(x, x + 3, ':r') # 红色实点线
```

## 4.3.2 调整图形：坐标轴上下限

```python
plt.plot(x, np.sin(x))
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
```

逆序坐标轴

```python
plt.plot(x, np.sin(x))

plt.xlim(10, 0)
plt.ylim(1.2, -1.2)
```

一行代码设置x，y的限制

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5])
```

还可以自动收紧坐标轴，不留空白区域。

```python
plt.plot(x, np.sin(x))
plt.axis('tight')
```

```python
plt.plot(x, np.sin(x))
```

让屏幕上显示的图形分辨率为1：1

```python
plt.axis('equal')
plt.plot(x, np.sin(x))
```

## 4.3.3 设置图形标签

图形标题和坐标轴标题

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
```

设置图例

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')
plt.legend()
```

在用面向对象接口画图时，采用ax.set()方法一次性设置所有的属性。

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim=(0, 10), ylim=(-2, 2), xlabel='x', ylabel='sin(x)', title='A Simple Plot')
# plt.axis('tight')
```
