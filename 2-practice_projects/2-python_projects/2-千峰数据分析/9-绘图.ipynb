```python
from pyecharts.charts import Bar
from pyecharts import options as opts 
import os

bar = Bar()
bar.add_xaxis(["衬衫", "毛衣", "领带", "裤子", "风衣", "高跟鞋", "袜子"])
bar.add_yaxis("商家A", [114, 55, 27, 101, 125, 27, 105], stack=True)
bar.add_yaxis("商家B", [57, 134, 137, 129, 145, 60, 49], stack=True)
bar.set_global_opts(title_opts=opts.TitleOpts(title="某商场销售情况"))
bar.render()
os.system("render.html")
```

```python
from pyecharts.charts import Map
value = [155, 10, 66, 78]
attr = ['福建', '山东', '北京', '上海']
map = Map(init_opts=opts.InitOpts(width=1200, height=600))
map.add('全国地图示例', [(i, j) for i, j in zip(attr, value)], maptype='china')
map.set_global_opts(
        title_opts=opts.TitleOpts(title="中国地图"),
        visualmap_opts=opts.VisualMapOpts(max_=300)
    )
map.render()
```

```python
x = np.linspace(0, 2*np.pi, 100)
plt.figure(figsize=(10,4))

ax = plt.subplot(1, 2, 1)
ax.plot(x, np.sin(x))
ax.set_title('Sin')

ax = plt.subplot(1, 2, 2)
ax.plot(x, np.cos(x))
ax.set_title('Cos')

plt.suptitle('Sine and Cosine Functions', fontsize=16, y=1.0)
```
