```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
```

```python
import sklearn.datasets as datasets
```

```python
iris = datasets.load_iris()
```

```python
X = iris['data']
```

```python
y = iris['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
import graphviz
```

```python
from sklearn import tree
```

```python
data = tree.export_graphviz(clf, out_file=None)
graph = graphviz.Source(data)
graph
```

```python
data = tree.export_graphviz(clf, out_file=None, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
graph = graphviz.Source(data)
graph
```

```python
from sklearn.datasets import load_wine
```

```python
wine = load_wine()
```

```python
wine
```

```python
X = wine['data']
y = wine['target']
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier, AdaBoostClassifier, GradientBoostingClassifier
```

```python
for model in [RandomForestClassifier, ExtraTreesClassifier, AdaBoostClassifier, GradientBoostingClassifier]:
    m = model()
    m.fit(X_train, y_train)
    print(m.score(X_test, y_test))
```

```python
xgb = XGBClassifier()
```

```python
xgb.fit(X_train, y_train)
```

```python
xgb.score(X_test, y_test)
```
